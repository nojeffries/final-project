## Analytic approach
I want to initially examine the variables by first considering their strength in a univariate logistic regression and viewing descriptive statistics. This should reveal some of the more important variables and odd results should be noted and examined. However the primary analyses will be multivariable - 1) Lasso logistic regression, and 2) random forest approaches. 


The continuous variables of interest are listed in a list called numvars.

Many of the potential explanatory variables have 4 or 5 ordinal categories indicated by number (e.g. 1, 2, 3) - as detailed in the data_dictionary.xlsx file the higher the number the better the rating/level. These categorical variables (e.g. education) need to handled/presented differently from continuous variables (e.g. age). I create a list of the names called facvars. I create new 0-1 dichotomous variables to convert these categorical varibles to this simpler dichotomous form that is easier to use in the lasso logistic regression.

First I convert *attrition* from a yes/no variable to a 1/0 as most logistic regression models use 1 and 0 as the dichotomous outcomes. 

```python
attrition_data['attrition'] = np.where(attrition_data['attrition']=="Yes",1,0) # reset attrition as a 0-1 variable

print(pd.crosstab(attrition_data['attrition'],columns="count",colnames=['']))

facvars=(['businesstravel','department','education','educationfield','gender','joblevel','jobrole',
         'maritalstatus','stockoptionlevel','jobinvolvement','performancerating','environmentsatisfaction',
         'jobsatisfaction', 'worklifebalance'])
print("\n\nfacvars = " + str(facvars))

numvars = (['age', 'monthlyincome', 'numcompaniesworked', 'percentsalaryhike',
           'totalworkingyears', 'yearsatcompany', 'yearssincelastpromotion', 'yearswithcurrmanager'] )
print('\n\nnumvars = ' + str(numvars))

```

Next I create the 0/1 simplifications for the categorical variables in **_facvars_**. These simplified variables are collected in a new list called **_newfacvars_**. An example of the recoding of an original variable is shown below. The recoding of the other variables is shown below that for categorical variables that showed some explantatory power in univariate regression.

```python
print("Original variable: businesstravel")
import matplotlib.pyplot as plt
attrition_data['businesstravel'].hist()
plt.show()
print(pd.crosstab(attrition_data['businesstravel'],columns="count",colnames=['']))
attrition_data['travelalot'] = np.where(attrition_data['businesstravel']=="Travel_Frequently",1,0)

print("\nNew variable: travelalot")
print(pd.crosstab(attrition_data['travelalot'],columns="count",colnames=['']))
```

```python
attrition_data['researchjob'] = np.where(attrition_data['jobrole'].isin(['Research Director','Research Scientist']),1,0)
attrition_data['evermarried'] = np.where(attrition_data['maritalstatus']=="Single",0,1)
attrition_data['highworkenvironment'] = np.where(attrition_data['environmentsatisfaction']>2,1,0)
attrition_data['highjobsatisfaction'] = np.where(attrition_data['jobsatisfaction']>2,1,0)
attrition_data['highworklifebalance'] = np.where(attrition_data['worklifebalance']>2,1,0)

newfacvars = (
['travelalot','researchjob','evermarried','highworkenvironment','highjobsatisfaction','highworklifebalance']
)
```

## Split into training and test sets
Machine learning approaches often split data into a training set that is used to develop prediction models, and a test set that is used to estimate the performance on a dataset that is unrelated to that used to develop the model. This approach is implemented below.


```python

jobs = attrition_data[['attrition']+numvars+newfacvars]

from sklearn.model_selection import train_test_split
X = jobs[numvars+newfacvars] # Features
y = jobs.attrition # Target variable
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=0)


train = X_train
train = train.assign(attrition=list(y_train))
print("\nTraining data head")
print(train.head())
print(train.shape)
```

### Show univariate logistic regressions for variables
I present some information about the univariate distributions for the explanatory variables and examine their strength in a univariate logistic regression with *attrition* as the dependent variable. These univariate analyses can give an idea of whether the variables have odd distributions that require further investigation and an opportunity to drop variables that fail to show some univariate association with *attrition*. All of these analyses are being conducted on the training dataset. 

Regression coefficients with positive(negative) values indicate characteristics that increase(decrease) the likelihood of leaving the company. I only include those variables with p-values < 0.10.

```python
from statsmodels.formula.api import logit
import statsmodels.api as sm

  
for var in numvars: 
    print('\n \n Univariate analysis of '+var)
    print(train[var].describe()) 
    form1 = 'attrition ~ '+var    
    fit = logit(form1,data=train).fit()
    print(fit.summary())

for var in newfacvars: 
    print('\n \n Univariate analysis of '+var)
    print(train[var].describe()) 
    form1 = 'attrition ~ '+var    
    fit = logit(form1,data=train).fit()
    print(fit.summary())

```

### Summary of univariate logistic regressions
Because the sample size is relatively large, it is easy to get highly significant results - 14 variables have p-values < 0.10.  Many of the results probably reflect the same issue(s) - e.g. yearsatcompany, yearswithcurrmanager both measure time at the company while totalworkingyears and age are similar. Similarly there are a number of factors related to job satisfaction. 

These 14 variables will be used in the multivariate lasso logistic regression and random forest approaches to prediction.


### Multivariable logistic regression with lasso 
The Lasso part of the model means the regression coefficients are constrained in magnitude. The regression coefficients $\beta_k, k=0, ..., 14$ 
are chosen so that the estimated probabilities of attrition, ($p_i$ for the $i^{th}$ person), fit the data well:
$$
p_i = \frac{exp\left(\beta_0+\beta_1 X_{i1} + ... + \beta_{14} X_{i14}\right)}{1+exp\left(\beta_0+\beta_1 X_{i1} + ... + \beta_{14} X_{i14}\right)}
\text{ and }\sum_{k=0}^{14} \left| \beta_k \right| < f(C).$$

Good values of $C$ are chosen through a cross-validation approach on the training set. In this example $C$ is chosen from 30 possible values. I can't really tell from the documentation exactly how $C$ constrains the magnitude of $\beta_k$ coefficients but qualitatively a smaller $C$ means $\beta_k$ coefficients forced toward zero in absolute value.

With the lasso approach the matrix of explanatory variables, $X$, is standardized so each variable has mean 0 and standard deviation of 1.

```python
from sklearn.linear_model import LogisticRegressionCV, LogisticRegression

X_tr = train[numvars+newfacvars]
y_tr = y_train
Xstan_tr = ((X_tr-X_tr.mean())/X_tr.std())
```

Output from the cross validation includes the estimated regression coefficients (clf.coef_), 
the $C$ value (clf.C_), and the estimated probabilities of attrition,
$$
p_i = \frac{exp\left(\beta_0+\beta_1 X_{i1} + ... + \beta_K X_{iK}\right)}{1+exp\left(\beta_0+\beta_1 X_{i1} + ... + \beta_K X_{iK}\right)}
$$
called **_predprobs_** here

```python
clf = LogisticRegressionCV(Cs=30,cv=20,random_state=0,penalty='l1',solver='liblinear').fit(Xstan_tr,y_tr)

#print(pd.DataFrame(clf.coef_.flatten(),index=Xstan_tr.columns,columns=['Coef']))
print("\nThe value of C is ",clf.C_,"\n")
predprobs = pd.DataFrame(clf.predict_proba(Xstan_tr)).iloc[:,1]

print("\nThe distribution of predicted values:\n")
print(predprobs.describe())
```

```python
print("\nThe predicted regression coefficients:\n")
from statsmodels.formula.api import logit
import statsmodels.api as sm
Xstan2 = sm.add_constant(Xstan_tr)

model = sm.Logit(y_tr, Xstan2)
result = model.fit_regularized(method='l1',alpha=clf.C_)
result.summary()
```


Important predictors that increase the likelihood of leaving are younger age, working for more previous companies, more years since last promotion, traveling more, being single, and having lower ratings of work environment/job satisfaction. 


16.1% of the training set individuals left the company. This suggests it is reasonable to predict attrition for the highest 16.1% predicted probabilities.

The cutpoint for predicting attrition is shown below, i.e. predict attrition if the person's predicted probability is above the cutpoint level of 0.28. 

```python
print(pd.crosstab(y_tr,columns="count",colnames=[''])/y_tr.count())

cutpoint = predprobs.quantile(1-.161)
print("\ncutpoint = ",cutpoint)
```

### Accuracy of multivariable logistic regression with lasso
Now check the accuracy of this rule on the test dataset (N=1075 individuals)

```python

test = X_test
test = test.assign(attrition=list(y_test))
print("\nShape of testing data")
print(test.shape)

X_te = test[numvars+newfacvars]
y_te = y_test
Xstan_te = ((X_te-X_te.mean())/X_te.std())

from sklearn.metrics import classification_report, confusion_matrix

predprobs = pd.DataFrame(clf.predict_proba(Xstan_te)).iloc[:,1]
zeroone = np.where(predprobs > cutpoint,1,0)

print("\nConfusion Matrix")
cm = confusion_matrix(y_te,zeroone,labels=[0,1])
print(cm)
```

```python
print("Number of 1075 where prediction=0 and truth=0: ",cm[0,0], ", percent = ",(cm[0,0]/1075).round(3))
print("Number of 1075 where prediction=0 and truth=1: ",cm[1,0], ", percent = ",(cm[1,0]/1075).round(3))
print("Number of 1075 where prediction=1 and truth=0: ",cm[0,1], ", percent = ",(cm[0,1]/1075).round(3))
print("Number of 1075 where prediction=1 and truth=1: ",cm[1,1], ", percent = ",(cm[1,1]/1075).round(3))

print("\nAccuracy = ", ((cm[0,0]+cm[1,1])/cm.sum()).round(3))

```

### Find accuracy of a Random Forest Approach

```python
from sklearn.ensemble import RandomForestClassifier

clf = RandomForestClassifier(n_estimators=50).fit(X_tr,y_tr)

y_pred2 = clf.predict(X_te)
cm_rf = confusion_matrix(y_te,y_pred2)
print(cm_rf)
```

```python
print("Number of 1075 where prediction=0 and truth=0: ",cm_rf[0,0], ", percent = ",(cm_rf[0,0]/1075).round(3))
print("Number of 1075 where prediction=0 and truth=1: ",cm_rf[1,0], ", percent = ",(cm_rf[1,0]/1075).round(3))
print("Number of 1075 where prediction=1 and truth=0: ",cm_rf[0,1], ", percent = ",(cm_rf[0,1]/1075).round(3))
print("Number of 1075 where prediction=1 and truth=1: ",cm_rf[1,1], ", percent = ",(cm_rf[1,1]/1075).round(3))

print("\nAccuracy = ", ((cm_rf[0,0]+cm_rf[1,1])/cm.sum()).round(3))

```

Show the most important variables using Random Forest approach

```python
print('Variable importance for Random Forest applied to training data set \n')
pd.DataFrame(clf.feature_importances_,columns = ['Importance'],
             index=X_tr.columns).sort_values(by=['Importance'],ascending=False)
```

### Conclusions
The random forest approach had much better prediction accuracy (98%) than the multivariable logistic regression with lasso constraint (80%). The important variables were quite different as well - monthly income was very important for the random forest but was one of the few insignificant variables in the logistic model. The superiority of the random forest is not so surprising - the more modern classification approaches frequently perform better than older parameteric regression approaches, but I was surprised at the extent of the difference.
